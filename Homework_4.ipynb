```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
data = pd.read_csv("kc-house-data.csv")
data.head()
```

```python
# Построение графика
plt.hist(data['price'], bins=500)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Цена')
plt.ylabel('Количество')
plt.show()


# Выводы
# График показывает (гамма) распределение стоимости недвижимости. 
# Мы видим, что большинство объектов имеют цену в диапазоне от 0 до 1 млн долларов, 
# при этом есть небольшое количество объектов с очень высокой стоимостью.
```

Большинство объектов имеют цену в диапазоне от 0 до 1 миллиона долларов, при этом есть небольшое количество объектов с очень высокой стоимостью.

# Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость <br> недвижимости, с применением не менее 5 диаграмм из урока. <br>
Анализ сделайте в формате storytelling: дополнить каждый график <br> письменными выводами и наблюдениями. <br> <br>

```python
import seaborn as sns

# Диаграмма 1: Точечный график цены в зависимости от квадратуры жилой площади
plt.scatter(data['sqft_living'], data['price'])
plt.title('Зависимость цены от квадратуры жилой площади')
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Цена')
plt.show()

# Выводы:
# Наблюдается положительная корреляция между квадратурой жилой площади и ценой недвижимости. 
# Чем больше площадь дома, тем выше его стоимость.

# Диаграмма 2: Ящик с усами для цены в зависимости от количества спален
sns.boxplot(x='bedrooms', y='price', data=data)
plt.title('Влияние количества спален на цену')
plt.xlabel('Количество спален')
plt.ylabel('Цена')
plt.show()

# Выводы:
# Наблюдается некоторое влияние количества спален на цену недвижимости. 
# Дома с большим количеством спален, как правило, имеют более высокую стоимость.

# Диаграмма 3: Гистограмма распределения цены в зависимости от наличия вид на набережную
sns.histplot(data=data, x='price', hue='waterfront', kde=True)
plt.title('Распределение цены в зависимости от вид на набережную')
plt.xlabel('Цена')
plt.ylabel('Количество')
plt.legend(['Нет', 'Да'])
plt.show()

# Выводы:
# Наблюдается существенное влияние наличия вид на набережную на цену недвижимости. 
# Дома с видом на набережную обычно имеют более высокую стоимость.

# Диаграмма 4: Корреляционная матрица
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Корреляционная матрица')
plt.show()

# Выводы:
# При анализе корреляционной матрицы видно, что цена недвижимости имеет сильную положительную корреляцию
# с площадью жилой площади, количеством спален и ванных комнат, а также с годом постройки.
# Эти характеристики имеют сильное влияние на цену недвижимости.
```
