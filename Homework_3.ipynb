```python
import pandas as pd

# Считываем данные с помощью pandas
df = pd.read_csv('kc-house-data.csv')

df.head()
```

## 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах <br> между годом реновацией дома и годом постройки дома. <br> Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df.apply(lambda row: row['yr_renovated'] - row['yr_built'] if row['yr_renovated'] != 0 else 0, axis=1)
df.head()
```

## 1.3 Создайте признаки года продажи, месяца продажи

```python
df['sale_year'] = df.apply(lambda row: pd.to_datetime(row['date']).year, axis=1)
df['sale_month'] = df.apply(lambda row: pd.to_datetime(row['date']).month, axis=1)
df.head()
```

## 1.4 Удалите признаки date, zipcode, lat, long

```python
df = df.drop(['date', 'zipcode', 'lat', 'long'], axis=1)
df.head()
```

# Задача 3

## 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms')['price'].mean().sort_values()
```

## 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition')['price'].agg(['min', 'mean', 'max'])
```

## 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index='waterfront', columns='view', aggfunc='size', fill_value=0)
```

## 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
df.groupby(['floors', 'bedrooms']).size().unstack().fillna(0)
```

## 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df.pivot_table(index='condition', columns='grade', values='price', aggfunc='median')
```
